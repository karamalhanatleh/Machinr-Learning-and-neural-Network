# ReLu activation function

```python
#import paskages
import numpy as np
import matplotlib.pyplot as plt
```

The ReLu activation function returns 0 if the input is negative otherwise return the input as it is.



###### f(x) = max(0,x)

```python
#
def relu(x):
    return np.maximum(0.0  ,  x)
```

```python
x= np.linspace(-10, 10, 50) 
x
```

```python
plt.plot(x,relu(x))
plt.xlabel("x") 
plt.ylabel("relu(x)")  
plt.show()
```
